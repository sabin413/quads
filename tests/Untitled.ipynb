```python
import pandas as pd

file1 = "~/JupyterLinks/nobackup/quads_data/GEOSFP/2024/01/2024-01-01.pkl"
file2 = "~/JupyterLinks/nobackup/quads_data/GEOSFP/2024/01/2024-01-02.pkl"
file3 = "~/JupyterLinks/nobackup/quads_data/GEOSFP/2024/01/two_merged_digest_GEOSFP_2024-01.pkl"

df1 = pd.read_pickle(file1)
df2 = pd.read_pickle(file2)
df3 = pd.read_pickle(file3)

```

```python
print(df1[0].keys())
```

```python
print(df1[10]["id_key"] == df3[10]["id_key"])
```

```python
ind = 8000

data1, data2, data3 = df1[ind]["quantiles"], df2[ind]["quantiles"], df3[ind]["quantiles"]
y1, x1 = data1[0], data1[1]
y2, x2 = data2[0], data2[1]
y3, x3 = data3[0], data3[1]

import matplotlib.pyplot as plt

plt.plot(x1, y1, label="df1")
plt.plot(x2, y2, label="df2")
plt.plot(x3, y3, label="df3")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title(f"Quantiles comparison (index={ind})")

plt.show()
```
